```python
# While Loop
# The while loop in Python is used to repeatedly execute a block of code as long as a specified condition is true.

count = 0
while count < 5:
    print("Count is:", count)
    count= count+1  # Increment the counter to avoid infinite loop
```

```python
count = 0
while count%2==0:
    print(count)
    count=count+1
```

```python
## Break
# The break statement exits the loop permanently.
for i in range(10):
    if i==5:
        break # Exit the loop when i is 5
    print(i)
```

```python
# Continue
for i in range(10):
    if i%2==0:
        continue # Skip even numbers (If the number is even it will skip it ==> Displaying all the odd numbers)
    print(i)
```

```python
# Pass
for i in range(5):
    if i ==3:
        pass  # Placeholder, does nothing 
    print(i)
```

```python
## Nested Loop
# A Loop inside another loop is called a nested loop.
for i in range(3):
    for j in range(2):
        print(f"i: {i}, j: {j}") # This is a f string
```

```python
# Examples : Calculate the sum of first n natural numbers using a while loop and for loop.

n=10
sum=0
count=1
while count<=n:
    sum=sum+count
    count = count+1

print("Sum of n natural numbers is", sum)
```

```python
# Now with for loop
sum = 0
for i in range(11):
    sum = sum+i
print(sum)
```
