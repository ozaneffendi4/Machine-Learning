```python
## Data Serilization
# Convert dictionary to json
import json
data = {'name': 'Ozan', 'age' : 17}

json_str = json.dumps(data)
print(json_str , type(json_str)) # Converted into string

# Converting back to dic
parsed_data = json.loads(json_str)
print(parsed_data , type(parsed_data)) # Converted into dict
```

```python
## CSV
import csv 
# open then mode = w (write)
with open('example.csv' , mode='w', newline="") as file:
    writer = csv.writer(file)
    writer.writerow(['name','ozan'])
    writer.writerow(['age' , 17]) # Create a example.csv file

with open('example.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
