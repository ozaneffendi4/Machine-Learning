```python
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        positive = abs(x)
        reversed = int(str(positive)[::-1])
        if x <0 :
            reversed = -reversed
        if not (-2**31 <= reversed <= 2**31 -1):
            return 0
        else:
            return reversed
```
