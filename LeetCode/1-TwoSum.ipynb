```python
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        result = []
        for i in range(len(nums)):
            for j in range(i+1 , len(nums)):
                if nums[i] + nums[j] == target:
                    result = [i,j]
        return result
```

```python
carry = 0
result = []

for i in range(len(rev1)):
    total = rev1[i] + rev2[i] + carry
    result.append(total % 10)
    carry = total // 10

if carry:
    result.append(carry)

print(result[::-1])
```
